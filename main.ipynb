```python
import random

DIFFICULTY = {
    '1': ('Easy',   10),
    '2': ('Medium',  7),
    '3': ('Hard',    5),
}

print('Guess the Number')
print('-' * 30)
print('Pick a difficulty:')
print('  1. Easy   — 10 tries')
print('  2. Medium —  7 tries')
print('  3. Hard   —  5 tries')

choice = input('\nEnter 1, 2 or 3: ').strip()

if choice not in DIFFICULTY:
    print('Invalid choice, defaulting to Medium')
    choice = '2'

level_name, max_tries = DIFFICULTY[choice]
print(f'\nLevel: {level_name} — you have {max_tries} tries. Good luck.')
```

```python
secret = random.randint(1, 100)
attempts = 0
won = False
hint_given = False

def get_hint(n):
    if n % 2 == 0:
        return 'Hint: the number is even.'
    else:
        return 'Hint: the number is odd.'

print(f'\nI have picked a number between 1 and 100.')
print(f'You have {max_tries} tries.\n')

while attempts < max_tries:
    remaining = max_tries - attempts

    try:
        guess = int(input(f'Attempt {attempts + 1}/{max_tries} — Enter your guess: '))
        if guess < 1 or guess > 100:
            raise ValueError('Number must be between 1 and 100')
    except ValueError as e:
        print(f'  Invalid input: {e}\n')
        continue

    attempts += 1

    if guess == secret:
        print(f'\nCorrect! The number was {secret}.')
        print(f'You got it in {attempts} attempt{"s" if attempts > 1 else ""}.')
        won = True
        break
    elif guess < secret:
        print(f'  Too low.  {remaining - 1} tries left.')
    else:
        print(f'  Too high. {remaining - 1} tries left.')

    if attempts == 3 and not hint_given:
        print(f'  {get_hint(secret)}')
        hint_given = True

    print()

if not won:
    print(f'\nOut of tries. The number was {secret}. Better luck next time.')
```
